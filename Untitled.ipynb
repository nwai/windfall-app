```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) load
df = pd.read_csv('windfall_history_lottolyzer.csv')
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')

# 2) 200×45 binary matrix (1 if ball drawn, else 0)
balls = range(1, 46)
presence = np.zeros((len(df), len(balls)), dtype=np.uint8)
for i, (_, row) in enumerate(df.iterrows()):
    drawn = row[['main1','main2','main3','main4','main5','main6','supp1','supp2']].values
    presence[i, drawn-1] = 1

# 3) colour: 0 = white, 1 = blue
cmap = plt.cm.colors.ListedColormap(['white', '#0059ff'])

# 4) plot
plt.figure(figsize=(10, 20))
plt.imshow(presence, cmap=cmap, aspect='auto', interpolation='nearest')
plt.xticks(range(0, 45), range(1, 46))
plt.yticks([])
plt.xlabel('Ball number')
plt.ylabel('Draws (oldest → newest)')
plt.title('Drawn (blue) vs not drawn (white)')
plt.tight_layout()
plt.show()
```
